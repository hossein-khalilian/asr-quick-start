```python
import os
import random
from datasets import load_from_disk, Audio
import IPython.display as ipd

home_dir = os.path.expanduser("~")
base_cache_path = os.path.join(home_dir, ".cache", "huggingface", "datasets", "common_voice")

def load_dataset_split(split):
    path = os.path.join(base_cache_path, f"commonvoice-fa-{split}")
    dataset = load_from_disk(path)
    dataset = dataset.cast_column("audio", Audio())
    return dataset

dataset = load_dataset_split("test")
print(f"Loaded 'dev' split with {len(dataset)} samples")

num_samples = 3
for _ in range(num_samples):
    idx = random.randint(0, len(dataset) - 1)
    sample = dataset[idx]
    print(f"Random sample index: {idx}")
    print(f"Text: {sample['sentence']}")  # adjust key if needed
    
    audio_data = sample["audio"]
    ipd.display(ipd.Audio(data=sample["audio"]["array"], rate=sample["audio"]["sampling_rate"]))
```
